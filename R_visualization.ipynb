```r
cone <- function(x, y){
sqrt(x ^ 2 + y ^ 2)
}
```

```r
x <- y <- seq(-1, 1, length = 30)
z <- outer(x, y, cone)
```

```r
persp(x, y, z,
main="Perspective Plot of a Cone",zlab = "Height",theta = 30, phi = 15,col = "orange", shade = 0.4)
```
